### Best Predictors

```python
def ensemble_feature_importances(df, labels, model, n = 10):
    """This is for `unstable` methods that generate different model each time they are called - ensemble tree methods"""
    importances = []
    for i in range(n):
        model.set_params(**{'random_state':i}) # change random state
        model.fit(df, labels) # train model
        importances.append(list(model.feature_importances_)) # append feature importances

    return np.mean(importances, axis = 0)

def rank_features(df, i):
    tempdf = pd.DataFrame(np.abs(df.iloc[:,i]).sort_values(ascending = False)) # sort coefficients
    tempdf['Ranking'] = np.arange(len(tempdf)) # rank them
    return np.array(tempdf.sort_index()['Ranking']) # return just the rank - and making sure that row indices in input and output match
```

```python
features_age_male = pd.DataFrame({
    'SVM':svm_mod.coef_[0],
    'Rdg':ridge_mod.coef_,
    'Las':lasso_mod.coef_,
    'PLS':pls_mod.coef_.T[0],
    'GrB':ensemble_feature_importances(train_df_male_stand, train_labels_male, grboost),
    'Ada':ensemble_feature_importances(train_df_male_stand, train_labels_male, adaboost),
    'XGB':ensemble_feature_importances(train_df_male_stand, train_labels_male, xgbmod),
    'RnF':ensemble_feature_importances(train_df_male_stand, train_labels_male, rfmod)
}, index = list(train_df_male))

features_age_male = features_age_male.sort_index()
features_age_male.head()
```

```python
def best_features(features_age):
    cordirr = np.empty((441, 3))

    for i in range(441):
        for j in range(3):
            if features_age.loc[:, ['SVM', 'Rdg', 'PLS']].iloc[i, j] > 0:
                cordirr[i, j] = 1
            else:
                cordirr[i, j] = -1

    correlation_dirrection = np.median(cordirr, axis = 1)
    correlation_dirrection = ['Positive' if i == 1 else 'Negative' for i in correlation_dirrection]
    
    rankings = []
    for i in range(features_age.shape[1]):
        rankings.append(rank_features(features_age, i))

    ranking_df = pd.DataFrame(np.array(rankings).T, columns = np.array(features_age.columns), index = np.array(features_age.index))
    ranking_df.head()
    
    best_features_age = pd.DataFrame({'Median_rank':np.median(ranking_df, axis=1),
                                        'Correlation':correlation_dirrection}, 
                                      index = ranking_df.index).sort_values('Median_rank',ascending = True)
    
    return(best_features_age)
```

```python
def best_features_plot(best_features_age, n = 15):
    best_features_age = best_features_age[:n]
    
    plt.figure(figsize=(10, 5))
    ax = sns.barplot(y = best_features_age.index, 
                x = best_features_age.Median_rank+1, 
                hue = best_features_age.Correlation, dodge = False, hue_order = ['Positive', 'Negative'],
                     palette = {'Positive':'C2', 'Negative':'orangered'}
               )

    plt.setp(ax.get_legend().get_texts(), fontsize='15')
    plt.setp(ax.get_legend().get_title(), fontsize='15')

    #plt.title('Best Features (Age Male) Median Rank', size = 15)
    plt.tick_params(labelsize = 15)
    plt.xlabel('Median Rank + 1', size = 15)
    plt.legend(fontsize = 15)
    plt.tight_layout()
```

```python
best_features_age_male = best_features(features_age_male)
best_features_age_male.head()
```

### Predicted values against actual values

```python
from sklearn.model_selection import cross_val_predict

def pred_vs_real(train_df, train_labels, test_df, test_labels, models, model_names, colors = ['C0', 'red'], cv = 10):
    
    fig, ax = plt.subplots(2, 4, figsize = (16, 8), sharex = True, sharey = True)

    for i, model in enumerate(models):
        train_preds = cross_val_predict(model, train_df, train_labels, cv = cv)
        test_preds = model.predict(test_df)
        ax[i // 4, i % 4].plot(np.arange(min(train_labels), max(train_labels)), np.arange(min(train_labels), max(train_labels)), '--k')
        ax[i // 4, i % 4].plot(train_labels, train_preds, '.', color = colors[0], label = 'CV', markersize = 10)
        ax[i // 4, i % 4].plot(test_labels, test_preds, '.', color = colors[1], label = 'Test', markersize = 10)
        ax[i // 4, i % 4].set_title(model_names[i], size = 15)
        ax[i // 4, i % 4].tick_params(labelsize = 15)
    ax[1, 3].legend(fontsize = 15)
    ax[0, 0].set_ylabel('Predicted Age', size = 15)
    ax[1, 0].set_ylabel('Predicted Age', size = 15)

    for i in range(4):
        ax[1, i].set_xlabel('Real Age', size = 15)

    plt.tight_layout()
```

```python
model_names = ['PLS', 'Ridge', 'Lasso', 'SVM', 'RF', 'GrBoost', 'AdaBoost', 'XGBoost']
```

```python
pred_vs_real(train_df_male_stand, train_labels_male, test_df_male_stand, test_labels_male, models, model_names, cv = 10)

plt.savefig('../../plots/Age_prediction/predicted_vs_real_male.png', dpi = 200)
```

### Residuals against Predicted values

```python
from sklearn.model_selection import cross_val_predict

def residual_plot(train_df, train_labels, test_df, test_labels, models, model_names, colors = ['C0', 'red'], cv = 10):
    fig, ax = plt.subplots(2, 4, figsize = (16, 8), sharex = True, sharey = True)
    minpred, maxpred = 100, 0
    for i, model in enumerate(models):
        train_preds = cross_val_predict(model, train_df, train_labels, cv = 10)
        test_preds = model.predict(test_df)

        if model_names[i] == 'PLS':
            residuals = train_preds.T[0] - np.array(train_labels)
            test_residuals = test_preds.T[0] - np.array(test_labels)
        else:
            residuals = train_preds - np.array(train_labels)
            test_residuals = test_preds - np.array(test_labels)

        standardized_residuals = (residuals - np.mean(residuals)) / np.std(residuals)
        standardized_test_residuals = (test_residuals - np.mean(test_residuals)) / np.std(test_residuals)

        maxpred = max(maxpred, max(train_preds))
        minpred = min(minpred, min(train_preds))

        ax[i // 4, i % 4].plot(train_preds, standardized_residuals, '.', color = colors[0], label = 'CV', markersize = 10)
        ax[i // 4, i % 4].plot(test_preds, standardized_test_residuals, '.', color = colors[1], label = 'Test', markersize = 10)
        ax[i // 4, i % 4].set_title(model_names[i], size = 15)
        ax[i // 4, i % 4].tick_params(labelsize = 15)

    ax[0, 0].set_ylabel('Normalized Residuals', size = 15)
    ax[1, 0].set_ylabel('Normalized Residuals', size = 15)

    ax[1, 3].legend(fontsize = 15)

    for i in range(4):
        ax[1, i].set_xlabel('Predicted Age', size = 15)

    for i in range(2):
        for j in range(4):
            ax[i, j].hlines(0, minpred, maxpred, ls = '--', alpha = 0.5)

    plt.tight_layout()
```

```python
residual_plot(train_df_male_stand, train_labels_male, test_df_male_stand, test_labels_male, models, model_names, cv = 10)

plt.savefig('../../plots/Age_prediction/residual_male.png', dpi = 200);
```

### Outlier Detection

**Elliptic Envelope**

Works well well $n_{samples} > n_{features} \cdot 2$ which is not this case

**Isolation Forests**

The ensemble.IsolationForest ‘isolates’ observations by randomly selecting a feature and then randomly selecting a split value between the maximum and minimum values of the selected feature.

Since recursive partitioning can be represented by a tree structure, the number of splittings required to isolate a sample is equivalent to the path length from the root node to the terminating node.

This path length, averaged over a forest of such random trees, is a measure of normality and our decision function.

Random partitioning produces noticeably shorter paths for anomalies. Hence, when a forest of random trees collectively produce shorter path lengths for particular samples, they are highly likely to be anomalies.

```python
from sklearn.ensemble import IsolationForest
```

```python
isoforest = IsolationForest(contamination='auto', behaviour='new')
isoforest.fit(train_df_male_stand)
```

**Local Outlier Factor**

The neighbors.LocalOutlierFactor (LOF) algorithm computes a score (called local outlier factor) reflecting the degree of abnormality of the observations. It measures the local density deviation of a given data point with respect to its neighbors. The idea is to detect the samples that have a substantially lower density than their neighbors.

In practice the local density is obtained from the k-nearest neighbors. The LOF score of an observation is equal to the ratio of the average local density of his k-nearest neighbors, and its own local density: a normal instance is expected to have a local density similar to that of its neighbors, while abnormal data are expected to have much smaller local density.

The number k of neighbors considered, (alias parameter n_neighbors) is typically chosen 1) greater than the minimum number of objects a cluster has to contain, so that other objects can be local outliers relative to this cluster, and 2) smaller than the maximum number of close by objects that can potentially be local outliers. In practice, such informations are generally not available, and taking n_neighbors=20 appears to work well in general. When the proportion of outliers is high (i.e. greater than 10 %, as in the example below), n_neighbors should be greater (n_neighbors=35 in the example below).

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
def meltbf(malebf, femalebf, n = 15, according_to = 'male'):
    
    if according_to == 'male':
        male = malebf[:n]
        female = femalebf.loc[male.index, :]        
    else:
        female = femalebf[:n]
        male = malebf.loc[female.index, :] 
    
    male['Sex'] = 'Male'
    female['Sex'] = 'Female'
    
    male.columns = ['Median_rank', 'Direction', 'Sex']
    female.columns = ['Median_rank', 'Direction', 'Sex']
    
    malefemale = pd.concat([male, female])
    rank_with_signs = [1+malefemale.iloc[i, :].Median_rank if malefemale.iloc[i, :].Direction == 'Positive' else -malefemale.iloc[i, :].Median_rank-1 for i in range(len(malefemale))]
    malefemale['Rank'] = rank_with_signs
    return malefemale
    #return male.head()
```

```python
mf = meltbf(best_features_age_male, best_features_age_female, according_to='male')
fm = malefemale = meltbf(best_features_age_male, best_features_age_female, according_to='female')
```
